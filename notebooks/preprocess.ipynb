```python
TARGET_SIZE = 800  # image resolution to be stored
MIN_SIZE_SHORTER = 256
IMG_QUALITY = 90  # JPG quality
```

```python
root = '/fs2/groups2/gca50080/yokoo/landmark/'
src_split = 'train2018_r1024'
dst_split = 'train2018_r800'

src_paths = set(map(lambda x: str(x).split('/')[-1].replace('.jpg', ''), Path(root + 'input/' + src_split).glob('*')))
dst_paths = set(map(lambda x: str(x).split('/')[-1].replace('.jpg', ''), Path(root + 'input/' + dst_split).glob('*')))
diff_ids = src_paths - dst_paths
```

```python
for i, id_ in tqdm(enumerate(diff_ids), total=len(diff_ids)):
    img = cv2.imread(f'../input/{src_split}/{id_}.jpg')
    
    height, width, _ = img.shape

    shorter_side = min(width, height)
    longer_side = max(width, height)

    if shorter_side < MIN_SIZE_SHORTER:
        continue

    if longer_side > TARGET_SIZE:
        factor = TARGET_SIZE / longer_side

        if round(shorter_side * factor) < MIN_SIZE_SHORTER:
            continue

        tw, th = round(width * factor), round(height * factor)
        img = img = cv2.resize(img, (tw, th))
    
    cv2.imwrite(f'../input/{dst_split}/{id_}.jpg', img)
```
